```python
from broadcaster_nets_2_3 import brGCN2_3

from torch_geometric.datasets import Planetoid
dataset = Planetoid(root='/tmp/Cora', name='Cora')

from typing import Optional, Tuple

import torch
from torch import Tensor
from torch.nn import Parameter
from torch_scatter import scatter_add
from torch_sparse import SparseTensor, fill_diag, matmul, mul
from torch_sparse import sum as sparsesum

from torch_geometric.nn.conv import MessagePassing
from torch_geometric.nn.dense.linear import Linear
from torch_geometric.nn.inits import zeros
from torch_geometric.typing import Adj, OptTensor, PairTensor
from torch_geometric.utils import add_remaining_self_loops, to_undirected
from torch_geometric.utils.num_nodes import maybe_num_nodes

from torch_geometric.loader import ClusterData
import numpy as np

import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv

import torch.nn as N

import itertools

outcomes = {}

iters = 3
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
data = dataset[0].to(device)

nets = []
number_layers = [2]
br_per_layer = [1,0]
output_per_layer = [4,2]
for layers in number_layers:
    br_combos_pre = list(itertools.product(br_per_layer, repeat = 1))
    br_combos_pre = [list(ele) for ele in br_combos_pre]
    br_combos = list(itertools.combinations_with_replacement(br_combos_pre, 2))
    output_combos = list(itertools.combinations_with_replacement(output_per_layer, (layers-1)))
    for output_combo in output_combos:
        for br_combo in br_combos:
            nets.append([layers, list(br_combo), list(output_combo)])
nets.reverse()

for idx in range(len(nets)):
    if nets[idx] == 0:
            continue
    outcome = {} 
    for runs in range(iters):
        model = brGCN2_3(data,dataset,nets[idx][0],nets[idx][1],nets[idx][2]).to(device)
        optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)

        for epoch in range(200):
            optimizer.zero_grad()
            out = model(data)
            loss = F.nll_loss(out[data.train_mask], data.y[data.train_mask])
            loss.backward()
            optimizer.step()

        model.eval()
        pred = model(data).argmax(dim=1)
        correct = (pred[data.test_mask] == data.y[data.test_mask]).sum()
        acc = int(correct) / int(data.test_mask.sum())
        print(f'loop:{runs} Accuracy: {acc:.4f}')
        outcome[runs] = acc

    print(f'average: {sum(outcome.values())/len(outcome.values())}')
    outcomes[model.name] = sum(outcome.values())/len(outcome.values())
    
    
for entry in outcomes:
    print(f'{entry} average: {outcomes[entry]}')
```
